```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.font_manager as fm

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])
y = np.array([5, 10, 11, 6])
plt.plot(xpoints, ypoints, 'o-C6', markersize = 5)
plt.plot(xpoints, y, color = 'black', marker = 'D', markersize = 5)
plt.xticks(range(10))
plt.yticks(range(13))
    
# 보조선
plt.grid(color = 'black', linestyle = '--', linewidth = 0.5)

plt.legend(['pink', 'black'])
plt.show()
```

```python
# plot1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x, y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(x, y)

plt.show()
```

```python
# plot1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.figure(figsize = (10, 10))
plt.subplot(2, 1, 1)
plt.title('Hot Pink')
plt.plot(x, y, color = 'hotpink')



plt.figure(figsize = (10, 10))
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 1, 2)
plt.title('navy')
plt.plot(x, y, 'navy')

plt.show()
```

# scatter 그래프는 두개의 값을 비교할수있다.

```python

N = 10
x = np.random.randint(10, size = N)
y = np.random.randint(10, size = N)

plt.scatter(x, y, c = 'hotpink')
plt.xticks(range(11))
plt.yticks(range(11))
plt.grid()


xx = np.random.randint(10, size = N)
yy = np.random.randint(10, size = N)

plt.scatter(xx, yy, c='navy')
plt.legend(['hotpink', 'navy'])
plt.title('Ex')
plt.show()

print(x)
print(y)
print(xx)
print(yy)
```

# bar : 막대차트

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8, 4, 1,])
c = np.array([2, 5, 3, 4])

plt.plot(x, c, color = 'C9')
plt.bar(x, y, color = 'purple')
plt.show()
```

```python

x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8, 4, 1])


# 옆으로 찍을때 ( 전치같은 느낌)
plt.barh(x, y, color = 'purple')
plt.show()
```

# pie : 원형차트

```python
y = np.array([22, 44, 11, 33])

plt.title('Pie', color = 'white', size = 30)
mylabels = ['aa','bb','cc','dd']


plt.pie(y, labels = mylabels, autopct = '%.f%%', textprops={'color' : 'white'})
plt.legend(title = 'ABCD', loc = 'center right')
plt.show()
```

```python
df = pd.read_csv('data/pop_seoul.csv')
df.info()
```

```python
df['year'].astype(str)  # year에 타입을 문자열로 바꾼다.
```

```python
plt.figure(figsize = (7, 5))  # 그래프 크기
plt.plot(df.year, df.population, color = 'hotpink', marker = 'o', linestyle = '--')
plt.title('Seoul_pop')
plt.xlabel('year')
plt.ylabel('population')

plt.show()
```

```python
plt.figure(figsize = (7, 5)) 
plt.bar(df.year, df.population, color = 'hotpink', width = 3)
plt.title('Seoul_pop')
plt.xlabel('year')
plt.ylabel('population')

plt.show()
```
