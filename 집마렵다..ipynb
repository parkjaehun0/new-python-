```python
weather = requests.get('https://www.weather.go.kr/w/obs-climate/land/past-obs/obs-by-day.do?stn=108&yy=2020&mm=1&obs=9')
html_doc = weather.text

soup = BeautifulSoup(html_doc, 'html.parser')


table = soup.find('table', {'class' : 'table-col table-cal'})
data = []
for i in table.find_all('tr'):
    tds = list(i.find_all('td'))
    
    for j in tds:
        if j.find('span'):
            aa = j.find('span').text
            data.append([aa])
            
data
    
ja = pd.Series(data)

df = pd.DataFrame({'1월': ja})
df 

    
```
